```python
import numpy as np 
import pandas as pd
```

```python
file_csv = [
    'base_info_SerieA.csv',
    'data_details_Aerial_SerieA.csv',
    'data_details_Assists_SerieA.csv',
    'data_details_Blocks_SerieA.csv',
    'data_details_Cards_SerieA.csv',
    'data_details_Clearances_SerieA.csv',
    'data_details_Dribbles_SerieA.csv',
    'data_details_Fouls_SerieA.csv',
    'data_details_Goals_SerieA.csv',
    'data_details_Interception_SerieA.csv',
    'data_details_Key passes_SerieA.csv',
    'data_details_Offsides_SerieA.csv',
    'data_details_Passes_SerieA.csv',
    'data_details_Possession loss_SerieA.csv',
    'data_details_Saves_SerieA.csv',
    'data_details_Shots_SerieA.csv',
    'data_details_Tackles_SerieA.csv',
    'data_summary_SerieA.csv'

]
data_list = []
for file in file_csv:
    feature = file.split('_')[2]
    data_temp = pd.read_csv(file)
    if file in ['base_info_SerieA.csv', 'data_summary_SerieA.csv']:
        data_list.append(data_temp)
        continue
    new_columns = {}
    for col in data_temp.columns:
        new_columns[col] = f"{col}_{feature}"
    data_temp.rename(columns= new_columns, inplace= True)

    data_list.append(data_temp)

data = pd.concat(data_list, axis = 1)
```

```python
data = data.drop(columns= 'Unnamed: 0')
```

```python
count_club = data['Club'].value_counts()
```

```python
valid_club = count_club[count_club > 20].index
data = data[data['Club'].isin(valid_club)]
```

```python
count_name = data['Name'].value_counts()
invalid_name = count_name[count_name > 1].index
invalid_name
```

```python
for name in invalid_name:
    print(data[data['Name'] == name])
```

```python
data.to_csv('data_semifinal_SerieA.csv')
```
