```python
import pandas as pd
import numpy as np
```

```python
file_csv = [
    'base_info_Ligue1.csv',
    'data_details_Aerial_Ligue1.csv',
    'data_details_Assists_Ligue1.csv',
    'data_details_Blocks_Ligue1.csv',
    'data_details_Cards_Ligue1.csv',
    'data_details_Clearances_Ligue1.csv',
    'data_details_Dribbles_Ligue1.csv',
    'data_details_Fouls_Ligue1.csv',
    'data_details_Goals_Ligue1.csv',
    'data_details_Interception_Ligue1.csv',
    'data_details_Key passes_Ligue1.csv',
    'data_details_Offsides_Ligue1.csv',
    'data_details_Passes_Ligue1.csv',
    'data_details_Possession loss_Ligue1.csv',
    'data_details_Saves_Ligue1.csv',
    'data_details_Shots_Ligue1.csv',
    'data_details_Tackles_Ligue1.csv',
    'data_summary_Ligue1.csv'

]
data_list = []
for file in file_csv:
    feature = file.split('_')[2]
    data_temp = pd.read_csv(file)
    if file in ['base_info_Ligue1.csv', 'data_summary_Ligue1.csv']:
        data_list.append(data_temp)
        continue
    new_columns = {}
    for col in data_temp.columns:
        new_columns[col] = f"{col}_{feature}"
    data_temp.rename(columns= new_columns, inplace= True)

    data_list.append(data_temp)

data = pd.concat(data_list, axis = 1)
```

```python
data = data.drop(columns= 'Unnamed: 0')
```

```python
count_club = data['Club'].value_counts()
```

```python
valid_club = count_club[count_club > 20].index
data = data[data['Club'].isin(valid_club)]
```
