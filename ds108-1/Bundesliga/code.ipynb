Detail Table

```python

# Click vào Detailed
click = driver.find_element(By.XPATH, "/html/body/div[4]/div[5]/ul/li[6]/a").click()
# Click vào nút first 
#click = driver.find_element(By.XPATH, "/html/body/div[4]/div[5]/div[2]/div[4]/div/dl[2]/dd[1]/a").click()

wait = WebDriverWait(driver, 10)
choose = driver.find_element(By.ID, 'category')
# 2. Dùng Select để tương tác
select = Select(choose)
data_details = pd.DataFrame()
# 3. Chọn theo giá trị value
option = choose.find_elements(By.TAG_NAME, 'option')
for o in option:
    data_name = 'data_details_' + o.text + '_Bundesliga'
    print(o.text)
    if o.text in ['Tackles','Interception', 'Fouls','Cards', 'Offsides', 'Clearances', 'Blocks', 'Saves' , 'Shots', 'Goals', 'Dribbles', 'Possession loss', 'Aerial', 'Passes', 'Key passes']:#, 'Assists']:
        print('Đã cào')
        continue
    select.select_by_value(o.get_attribute('value'))
    wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[4]/div[5]/div[6]/div[3]/div/table/thead")))
    wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[4]/div[5]/div[6]/div[3]/div/table/tbody")))  # Đợi tbody xuất hiện
    col_name = []
    head = driver.find_element(By.XPATH, "/html/body/div[4]/div[5]/div[6]/div[3]/div/table/thead")
    titles = head.find_elements(By.TAG_NAME, "th")
    for t in titles:
        col_name.append(t.text)
    col_del = ['', 'Player', 'Apps', 'Mins', 'Rating']
    for c in col_del:
        col_name.remove(c)
    data_details = pd.DataFrame(columns= col_name)
    
    wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[4]/div[5]/div[6]/div[3]/div/table/tbody")))
    wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[4]/div[5]/div[6]/div[4]/div/dl[2]/dd[3]/a')))
    while True:
        try:
            next_button = driver.find_element(By.XPATH, '/html/body/div[4]/div[5]/div[6]/div[4]/div/dl[2]/dd[3]/a')
            if 'disabled' in next_button.get_attribute('class'):
                time.sleep(2)
            if 'disabled' in next_button.get_attribute('class'):
                tbody = driver.find_element(By.XPATH, '/html/body/div[4]/div[5]/div[6]/div[3]/div/table/tbody')
                rows = tbody.find_elements(By.TAG_NAME, 'tr')
                for r in rows:
                    each_row = []
                    td = r.find_elements(By.TAG_NAME, 'td')
                    for t in td:
                        each_row.append(t.text)
                    each_row = each_row[4:-1]
                    data_details.loc[len(data_details)] = each_row
                print("Nút Next bị disable rồi.")
                break
            else:
                tbody = driver.find_element(By.XPATH, '/html/body/div[4]/div[5]/div[6]/div[3]/div/table/tbody')
                rows = tbody.find_elements(By.TAG_NAME, 'tr')
                for r in rows:
                    each_row = []
                    td = r.find_elements(By.TAG_NAME, 'td')
                    for t in td:
                        each_row.append(t.text)
                    each_row = each_row[4:-1]
                    data_details.loc[len(data_details)] = each_row
                
                next_button.click()
                wait.until(EC.staleness_of(tbody))
                wait.until(EC.presence_of_element_located((By.XPATH, "/html/body/div[4]/div[5]/div[6]/div[3]/div/table/tbody")))
                

        except NoSuchElementException:
            print("Không tìm thấy nút Next.")
            break
    data_details.to_csv(data_name + '.csv', index = False)
        
        
    
```
