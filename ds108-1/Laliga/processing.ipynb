```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
file_csv = [
    'base_info_Laliga.csv',
    'data_details_Aerial_Laliga.csv',
    'data_details_Assists_Laliga.csv',
    'data_details_Blocks_Laliga.csv',
    'data_details_Cards_Laliga.csv',
    'data_details_Clearances_Laliga.csv',
    'data_details_Dribbles_Laliga.csv',
    'data_details_Fouls_Laliga.csv',
    'data_details_Goals_Laliga.csv',
    'data_details_Interception_Laliga.csv',
    'data_details_Key passes_Laliga.csv',
    'data_details_Offsides_Laliga.csv',
    'data_details_Passes_Laliga.csv',
    'data_details_Possession loss_Laliga.csv',
    'data_details_Saves_Laliga.csv',
    'data_details_Shots_Laliga.csv',
    'data_details_Tackles_Laliga.csv',
    'data_summary_Laliga.csv'

]
data_list = []
for file in file_csv:
    feature = file.split('_')[2]
    data_temp = pd.read_csv(file)
    if file in ['base_info_Laliga.csv', 'data_summary_Laliga.csv']:
        data_list.append(data_temp)
        continue
    new_columns = {}
    for col in data_temp.columns:
        new_columns[col] = f"{col}_{feature}"
    data_temp.rename(columns= new_columns, inplace= True)

    data_list.append(data_temp)

data = pd.concat(data_list, axis = 1)
```

```python
data = data.drop(columns= 'Unnamed: 0')
```
