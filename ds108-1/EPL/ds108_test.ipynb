Đổi tên thuộc tính 

```python
file_csv = [
    'base_info_EPL.csv',
    'data_details_Aerial_EPL.csv',
    'data_details_Assists_EPL.csv',
    'data_details_Blocks_EPL.csv',
    'data_details_Cards_EPL.csv',
    'data_details_Clearances_EPL.csv',
    'data_details_Dribbles_EPL.csv',
    'data_details_Fouls_EPL.csv',
    'data_details_Goals_EPL.csv',
    'data_details_Interception_EPL.csv',
    'data_details_Key passes_EPL.csv',
    'data_details_Offsides_EPL.csv',
    'data_details_Passes_EPL.csv',
    'data_details_Possession loss_EPL.csv',
    'data_details_Saves_EPL.csv',
    'data_details_Shots_EPL.csv',
    'data_details_Tackles_EPL.csv',
    'data_summary_EPL.csv'
]
data_list = []
for file in file_csv:
    # Lấy tên đặc trưng của từng file : Aerial, Assists, Blocks, Cards, ...
    feature = file.split('_')[2]
    data_temp = pd.read_csv(file)
    if file in [ 'base_info_EPL.csv', 'data_summary_EPL.csv']:
        data_list.append(data_temp)
        continue

    new_columns = {}
    for col in data_temp.columns:
        new_columns[col] = f"{col}_{feature}"
    data_temp.rename(columns= new_columns, inplace= True)

    data_list.append(data_temp)


data = pd.concat(data_list, axis = 1)


```

```python
data.columns
```

```python
for col in data.columns:
    num_null = data[col].isna().sum()
    print(col, num_null)
```

Vấn đề missing value sẽ được giải quyết sau. Bước hiện tại chỉ xử lý các vấn đề :
- Trùng lặp các cầu thủ (một mùa giải cầu thủ có thể mang cho mượn ở các clb khác nhau)
- Một số đội bóng đã xuống hạng nhưng vẫn được thêm vào (dữ liệu có sai sót)
