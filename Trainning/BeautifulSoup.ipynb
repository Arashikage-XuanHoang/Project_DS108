# Web Scraping with BeautifulSoup and Pandas

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
r = requests.get(url)

soup = BeautifulSoup(r.text, 'lxml')
```

```python
name = soup.find_all('a', class_ = 'title')

for n in range(len(name)):
  name[n] = name[n].string

price = soup.find_all("h4", class_ = 'price float-end card-title pull-right')

for p in range(len(price)):
  price[p] = price[p].text


des = soup.find_all("p", class_ = 'description card-text')
for d in range(len(des)):
  des[d] = des[d].string

reviews = soup.find_all('p', class_ = "review-count float-end")
for r in range(len(reviews)):
  reviews[r] = reviews[r].text




```

```python
data = pd.DataFrame()

data['Prodeuct Name']  = name
data['Price'] = price
data['Description'] = des
data['Reviews Count'] = reviews


data.to_csv("Products_details.csv")
```

# How to Scrape a Table From a Website using BeautifulSoup

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://ticker.finology.in/"
r = requests.get(url)
```

```python
print(r.status_code)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
table = soup.find_all('table', class_="table table-sm table-hover screenertable")[0]
```

```python

titles = []
headers = table.find_all("th")
for h in headers:
  print(h.text)
  titles.append(h.text)
```

```python
data_df = pd.DataFrame(columns= titles)
data_df.info()
```

```python
data = table.find('tbody')
rows = data.find_all('tr')
for r in rows:
  name = r.find_all('td')[0].text
  name = name.replace('\n', '').strip()
  price = r.find_all('td')[1].text
  high_price = r.find_all('td')[2].text
  data_df.loc[len(data_df)] = [name, price, high_price]
```

## Phía trên ta mới chỉ đang tìm page 1. Nhưng trang web lại có tận 50 trang.
idea : Can thiệp vào url, chạy một vòng for vào url để sửa page{i}

```python
def printt(r):
  soup = BeautifulSoup(r.content, 'lxml')
  li = soup.find_all('li', class_ = "col-xs-6 col-sm-4 col-md-3 col-lg-3")
  for l in li:
    a = l.find('img')
    print(a.attrs['alt'])
    star = l.find('p')
    print(star.attrs['class'][1])

```

```python
for i in range(1, 51):
  url = f"https://books.toscrape.com/catalogue/page-{i}.html"
  r = requests.get(url)
  printt(r)
```

## Một cách khác mà không cần dùng vòng for, một số trường hợp url sẽ không có trọng số chứa số trang

```python
import requests as re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://books.toscrape.com/catalogue/page-1.html"
```

```python

while True:
   r = re.get(url)
   soup = BeautifulSoup(r.text, 'html.parser')
   next = soup.find('li', class_ = 'next')

   if next:  # Nếu tìm thấy phần tử <li class="next">
      next_link = next.find('a').attrs['href']
      url = 'https://books.toscrape.com/catalogue/' + next_link
      print(next_link, f" Url tương ứng : f{url}")
   else:
      print("Không tìm thấy nút Next trên trang này.")
      break

```
