```python
import pandas as pd 
import numpy as np
```

```python
file_csv = [
    '1.FC Heidenheim 1846.csv',
    '1.FC Union Berlin.csv',
    '1.FSV Mainz 05.csv',
    'Bayer 04 Leverkusen.csv',
    'Bayern Munich.csv',
    'Borussia Dortmund.csv',
    'Borussia Mönchengladbach.csv',
    'Eintracht Frankfurt.csv',
    'FC Augsburg.csv',
    'FC St. Pauli.csv',
    'Holstein Kiel.csv',
    'RB Leipzig.csv',
    'SC Freiburg.csv',
    'SV Werder Bremen.csv',
    'TSG 1899 Hoffenheim.csv',
    'VfB Stuttgart.csv',
    'VfL Bochum.csv',
    'VfL Wolfsburg.csv'
]
```

```python
all_data = []
for file in file_csv:
    df = pd.read_csv(file)
    all_data.append(df)

data = pd.concat(all_data, ignore_index= True)
data = data.drop(columns= 'Unnamed: 0')
```

```python
data.to_csv('Bundesliga_Value.csv')
```
