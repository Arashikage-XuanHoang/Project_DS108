```python
import pandas as pd
file_csv = [
    'AC Milan.csv', 'AC Monza.csv', 'ACF Fiorentina.csv', 'AS Roma.csv',
    'Atalanta BC.csv', 'Bologna FC 1909.csv', 'Cagliari Calcio.csv',
    'Como 1907.csv', 'FC Empoli.csv', 'Genoa CFC.csv', 'Hellas Verona.csv',
    'Inter Milan.csv', 'Juventus FC.csv', 'Parma Calcio 1913.csv',
    'SS Lazio.csv', 'SSC Napoli.csv', 'Torino FC.csv',
    'Udinese Calcio.csv', 'US Lecce.csv', 'Venezia FC.csv'
]
all_data = []
for file in file_csv:
    df = pd.read_csv(file)
    all_data.append(df)

data = pd.concat(all_data, ignore_index= True)
data = data.drop(columns= 'Unnamed: 0')
data.to_csv('SerieA_Value.csv')
```
