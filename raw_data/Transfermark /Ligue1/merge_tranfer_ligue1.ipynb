```python
import pandas as pd
file_csv = [
    'AJ Auxerre.csv', 'Angers SCO.csv', 'AS Monaco.csv', 'AS Saint-Étienne.csv',
    'FC Nantes.csv', 'FC Toulouse.csv', 'Le Havre AC.csv', 'LOSC Lille.csv',
    'Montpellier HSC.csv', 'OGC Nice.csv', 'Olympique Lyon.csv', 'Olympique Marseille.csv',
    'Paris Saint-Germain.csv', 'RC Lens.csv', 'RC Strasbourg Alsace.csv',
    'Stade Brestois 29.csv', 'Stade Reims.csv', 'Stade Rennais FC.csv'
]
all_data = []
for file in file_csv:
    df = pd.read_csv(file)
    all_data.append(df)

data = pd.concat(all_data, ignore_index= True)
data = data.drop(columns= 'Unnamed: 0')
data.to_csv('Ligue1_Value.csv')
```
