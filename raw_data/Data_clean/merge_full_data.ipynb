```python
import pandas as pd   
import numpy as np 
```

```python
data_list = [
    'Bundesliga_ver1.csv',
    'EPL_ver1.csv',
    'Laliga_ver1.csv',
    'Ligue1_ver1.csv',
    'SerieA_ver1.csv'
]

data = pd.DataFrame()
for d in data_list:
    df = pd.read_csv(d)
    data = pd.concat([data, df], axis = 0, ignore_index= True )
```
