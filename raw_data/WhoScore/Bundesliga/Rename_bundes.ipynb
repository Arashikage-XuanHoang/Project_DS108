```python
list_group_att = ['Aerial', 'Assists', 'Blocks', 'Cards', 'Clearances', 'Dribbles', 'Fouls', 'Goals', 'Interception', 'Key passes', 'Offsides', 'Passes', 'Possession loss', 'Saves', 'Shots', 'Tackles']    
```

```python
for att in list_group_att:
    data = pd.read_csv('data_details_' + att + '_Bundesliga.csv')
    '''
    if att == 'Aerial':
        data.rename({
            'Total' : 'TotAD', 
            'Won' : 'WonAD',
            'Lost' : 'LostAD'
        }, inplace= True)
    '''
    if att == 'Assists':
        data.rename(columns= {
            'Cross' : 'CrAss',
            'Corner' : 'CorAss',
            'Throughball' : 'ThrbAss',
            'Freekick' : 'FreAss',
            'Throwin' : 'ThrInAss',
            'Other' : 'OthAss',
            'Total' : 'TotAss'
        }, inplace= True)
    elif att == 'Blocks':
        data.rename(columns={
            'ShotsBlocked' : 'BlkSh',
            'CrossesBlocked' : 'BlkCr',
            'PassesBlocked' : 'BlkPs'
        }, inplace = True)
    elif att == 'Cards':
        data.rename(columns={
            'Yellow' : 'YelC',
            'Red' : 'RedC'
        }, inplace= True)
    elif att == 'Clearances':
        data.rename(columns={
            'Total' : 'TotClr'
        }, inplace= True)
    elif att == 'Dribbles':
        data.rename(columns={
            'Unsuccessful' : 'UnDrib',
            'Successful' : 'SucDrib',
            'Total Dribbles' : 'TotDrib'
        }, inplace= True)
    elif att == 'Goals':
        data.rename(columns={
            'Total' : 'TotGs',
            'SixYardBox' : 'SYBGs',
            'PenaltyArea' : 'PAGs',
            'OutOfBox' : 'OOBGs'
        }, inplace= True)
    elif att == 'Interception':
        data.rename(columns={
            'Total' : 'TotInt'
        }, inplace= True)
    elif att == 'Key passes':
        data.rename(columns={
            'Total' : 'TotKPs',
            'Long' : 'LKPs',
            'Short' : 'SKPs'
        }, inplace= True)
    elif att == 'Offsides':
        data.rename(columns={
            'CaughtOffside' : 'COF'
        }, inplace= True)
    elif att == 'Passes':
        data.rename(columns={
            'Total' : 'TotPs'
        }, inplace = True)
    elif att == 'Possession loss':
        data.rename(columns={
            'UnsuccessfulTouches' : 'UnTch',
            'Dispossessed' : 'Dispo'

        }, inplace= True)
    elif att == 'Saves':
        data.rename(columns={
            'Total' : 'TotSav',
            'SixYardBox' : 'SYBSav',
            'PenaltyArea' : 'PASav',
            'OutOfBox' : 'OOBSav'

        }, inplace= True)
    elif att == 'Shots':
        data.rename(columns={
            'Total' : 'TotSh',
            'OutOfBox' : 'OOBSh',
            'SixYardBox' : 'SYBSh',
            'PenaltyArea' : 'PASh'

        }, inplace= True)
    elif att == 'Tackles':
        data.rename(columns={
            'TotalTackles' : 'TotTkl',
            'DribbledPast' : 'DribPast',
            'TotalAttemptedTackles' : 'TotAttTkl'

        }, inplace= True)
    data.to_csv('data_details_' + att + '_Bundesliga.csv')

```
