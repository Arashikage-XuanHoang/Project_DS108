```python
import pandas as pd
file_csv = [
    'Athletic Bilbao.csv', 'Atlético de Madrid.csv', 'CA Osasuna.csv', 'CD Leganés.csv',
    'Celta de Vigo.csv', 'Deportivo Alavés.csv', 'FC Barcelona.csv', 'Getafe CF.csv',
    'Girona FC.csv', 'Rayo Vallecano.csv', 'RCD Espanyol Barcelona.csv', 'RCD Mallorca.csv',
    'Real Betis Balompié.csv', 'Real Madrid.csv', 'Real Sociedad.csv',
    'Real Valladolid CF.csv', 'Sevilla FC.csv', 'UD Las Palmas.csv',
    'Valencia CF.csv', 'Villarreal CF.csv'
]
all_data = []
for file in file_csv:
    df = pd.read_csv(file)
    all_data.append(df)

data = pd.concat(all_data, ignore_index= True)
data = data.drop(columns= 'Unnamed: 0')
data.to_csv('Laliga_Value.csv')
```
