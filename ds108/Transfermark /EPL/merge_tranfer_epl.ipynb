```python
import pandas as pd
```

```python
file_csv = [
    'AFC Bournemouth.csv', 'Arsenal FC.csv', 'Aston Villa.csv', 'Brentford FC.csv',
    'Brighton & Hove Albion.csv', 'Chelsea FC.csv', 'Crystal Palace.csv', 'Everton FC.csv',
    'Fulham FC.csv', 'Ipswich Town.csv', 'Leicester City.csv', 'Liverpool FC.csv',
    'Manchester City.csv', 'Manchester United.csv', 'Newcastle United.csv',
    'Nottingham Forest.csv', 'Southampton FC.csv', 'Tottenham Hotspur.csv',
    'West Ham United.csv', 'Wolverhampton Wanderers.csv'
]
all_data = []
for file in file_csv:
    df = pd.read_csv(file)
    all_data.append(df)

data = pd.concat(all_data, ignore_index= True)
data = data.drop(columns= 'Unnamed: 0')
data.to_csv('EPL_Value.csv')
```
